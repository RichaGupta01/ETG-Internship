# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("summer.csv")
df
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
print(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
f = df[(df['Medal'] == 'Gold') ]
data=[]
for brand in f['Sport'].unique():
    data.append([brand , len(f[f['Sport']  == brand])])
print(data)    

pd.DataFrame(data,columns = ['brand','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'brand', y = 'freq', kind = 'bar', figsize = (20,20))
```
